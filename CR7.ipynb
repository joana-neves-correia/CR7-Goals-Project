## Coluna Season
### Uniformização dos valores inseridos como temporadas

```python
#verificar quais os valores existentes na coluna season
df.Season.unique()
```

```python
#substiruir os valores não conformes 
df.Season = df.Season.replace({'Dec-13':'12/13'})
df.Season.unique()
```

```python
# Função para normalizar épocas no formato YYYY/YYYY
def to_season_full(season):
    season = str(season).strip()
    
    if "/" in season and len(season) == 5:
        start, end = season.split("/")
        start_year = int("20" + start)   # 2002
        end_year = int("20" + end)       # 2003
        return f"{start_year}/{end_year}"
    return season

# Aplicar normalização
df["Season"] = df["Season"].apply(to_season_full).astype(str) 
```

# Coluna Result com resultados 'clean'

```python
df["Result_clean"] = df["Result"].str.replace(" AET", "")
df["Result_clean"] = df["Result_clean"].str.replace(" on pens", "")
df.Result_clean.unique()
```

### Colunas com os golos da equipa e do adversário

```python
# Separar os golos antes e depois dos :
df[["Goals1", "Goals2"]] = df["Result_clean"].str.split(":", expand=True)

# Converter para números
df["Goals1"] = df["Goals1"].astype(int)
df["Goals2"] = df["Goals2"].astype(int)

# Agora criar as colunas finais com base no Venue
df["Goals_Team"] = df.apply(
    lambda row: row["Goals1"] if row["Venue"] == "H" else row["Goals2"],
    axis=1
)

df["Goals_Opponent"] = df.apply(
    lambda row: row["Goals2"] if row["Venue"] == "H" else row["Goals1"],
    axis=1
)
```

```python
df[['Venue','Result','Goals1','Goals2','Goals_Team','Goals_Opponent']].head(10)
```

## Nova coluna com o momento de decisão (Penalties, Tempo extra, normal)

```python
def decision(x):
    if "AET" in x:
        return "After Extra Time"
    elif "on pens" in x:
        return "Penalties"
    else:
        return "Normal"

df["Decision"] = df["Result"].apply(decision)
```
